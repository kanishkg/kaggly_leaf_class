```python
import numpy as np
import os
import scipy as sp
import scipy.ndimage as ndi
from skimage import io
from skimage import measure
import cv2
```

```python
folder = 'images_pre'
image_width = 500
image_height = 500
image_files = os.listdir(folder)
patch_size = 8
num_samples = 20000
num_images = 1000
patch_folder = 'patches/'
num_saved = 0
```

```python
def get_contour(img):
    """returns the coords of the longest contour"""
    return max(measure.find_contours(img, .8), key=len)


def downsample_contour(coords, bins=512):
    """splits the array to ~equal bins, and returns one point per bin"""
    edges = np.linspace(0, coords.shape[0], 
                       num=bins).astype(int)
    for b in range(bins-1):
        yield [coords[edges[b]:edges[b+1],0].mean(), 
               coords[edges[b]:edges[b+1],1].mean()]
```

```python
def save_patches(image_data,cnt):
    global num_saved
    if len(cnt)<21:
        return
    imgs = range(1, len(cnt))
    num = np.random.choice(imgs, size=num_samples/num_images, replace=False)
    for i in num:
        x = cnt[i][0][0]
        y = cnt[i][0][1]
        patch = image_data[x-patch_size/2:x-1+patch_size/2,y-patch_size/2:y-1+patch_size/2]
        io.imsave(patch_folder+str(num_saved)+'.jpg', patch)
        num_saved+=1
```

```python
for image_index, image in enumerate(image_files[:num_images]):
  image_file = os.path.join(folder, image)
  print image_file
  image_data = cv2.imread(image_file)
  imgray = cv2.cvtColor(image_data,cv2.COLOR_BGR2GRAY)
  ret,thresh = cv2.threshold(imgray,127,255,0)
  im2, cnt, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
  save_patches(image_data,cnt[0])
  
```
